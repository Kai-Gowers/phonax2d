```python
import ase
from ase import Atoms
from phonax.phonons import atoms_to_ext_graph
```

```python
atoms = Atoms(
    symbols='MoTe2',
    positions=[(0,0,0), (1.8,1.0,0), (0.9,2.0,0)], 
    cell=[[3.5,0,0],[0,3.5,0],[0,0,20.0]],        
    pbc=[True, True, False]
)
```

```python
print("PBC flags:", atoms.pbc)
print("Cell:\n", atoms.cell)
```

```python
from ase.neighborlist import neighbor_list
i, j, S = neighbor_list("ijS", atoms, cutoff=3.0)  
print("Unique S z-components:", sorted({s[2] for s in S}))
```

```python
g = atoms_to_ext_graph(atoms, cutoff=3.0, num_message_passing=2)
```
